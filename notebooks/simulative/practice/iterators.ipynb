```python
class DataStreamIterator:
    """Кастомный итератор для ленивой генерации диапазона чисел."""
    def __init__(self, limit: int):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        # Соответствие протоколу: итератор возвращает сам себя
        return self

    def __next__(self):
        if self.current >= self.limit:
            # Единственный правильный способ завершить итерацию
            raise StopIteration
        
        result = self.current
        self.current += 1
        return result

# Использование
stream = DataStreamIterator(limit=3)
for value in stream:
    print(value)
```

#### генераторы

функция возвращает значение с помощью оператора `yield`
обычная функция отрабатывает после запуска полностью, генераторы делают вычисления когда достигли `yield` - возращает значение и останавливается в этом места а после очередного вызова продолжает с остановленного места

генератор является итератором
но есть разница
задача итератора - вернуть очередной элемент или бросить исключение
генератор создает элемент на лету (посчитал - выкинул из памяти и так дальше)

```python
def pow(x: int):
    while True:
        yield x ** 2
        x += 1
```

```python
p = pow(2)
```

```python
for l in p:
    print(l)
    break
```

```python
def fib():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
```

```python
f = fib()
```

```python
next(f)
```

```python
for i in range(20):
    print(next(f))
```

```python
def card_deck():
    m = ["Пик", "Трефы", "Крестей", "Бубны"]
    numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10, "В", "Д", "К", "A"]

    for i in m:
        for number in numbers:
            yield f"{i} {number}"

for card in card_deck():
    print(card)
```

```python
def show_let(s):
    for l in s:
        yield l
```

```python
res = show_let('Python is a popular programming language')

print(next(res))
# P

print(next(res))
```

```python
def substring_partitions(string: str, n: int):
    for i in range(1, min(n + 1, len(string) + 1)):
        if i == len(string):
            yield (string,)
        
        for partition in substring_partitions(string[i:], n):
            yield (string[:i],) + partition
```

```python
list(substring_partitions('abcd', 3))
```
