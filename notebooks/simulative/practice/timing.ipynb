Напишите функцию most_frequent_weekday, которая принимает список дат в формате год-месяц-день и возвращает самый часто встречающийся день недели в этом списке (например, "Monday").

Если несколько дней встречаются одинаково часто, функция должна вернуть все эти дни.

Пример

most_frequent_weekday(['2023-06-01', '2023-06-02', '2023-06-09'])
['Friday']

most_frequent_weekday(['2023-01-01', '2023-05-01', '2023-12-25'])

['Monday']

most_frequent_weekday(['2023-02-28', '2023-03-01', '2023-03-07', '2023-03-08'])
['Tuesday', 'Wednesday']

```python
from datetime import datetime

def most_frequent_weekday(dates: list[str]):
    dts = {}
    for date_str in dates:
        date_ =  datetime.strptime(date_str, "%Y-%m-%d").date()
        day_name = datetime.strftime(date_, "%A")
        dts[day_name] = dts.get(day_name, 0) + 1
    res = []
    largest_counter = max(dts.values())
    for day_name, counter in dts.items():
        if counter == largest_counter:
            res.append(day_name)
    return res

most_frequent_weekday(["2023-02-28", "2023-03-01", "2023-03-07", "2023-03-08"])
```

Напишите функцию get_weekdays, которая принимает список дат в формате год-месяц-день и возвращает список дней недели для каждой даты.

Пример

get_weekdays(['2023-06-01', '2023-06-02', '2023-06-03'])

['Thursday', 'Friday', 'Saturday']

get_weekdays(['2023-04-05'])

['Wednesday']

```python
from datetime import date

def get_weekdays(dates: list[str]):
    dates = [date.strptime(date_str, "%Y-%m-%d") for date_str in dates]
    return [date.strftime(date_, "%A") for date_ in dates]
get_weekdays(["2023-06-01", "2023-06-02", "2023-06-03"])
get_weekdays(["2023-04-05"])
```

При проведении разного рода исследований нам может потребоваться проанализировать разницу во времени между последовательными датами. Такой анализ может помочь нам понять временные промежутки между событиями и принимать решения на основе этого.

Напишите функцию, которая принимает список строк в качестве входных данных и возвращает среднюю разницу во времени между последовательными датами в секундах.

Пример

avg_time_diff(['2023-05-28 12:00:00'])

0.0

avg_time_diff(['2023-05-28 12:00:00', '2023-05-28 13:00:00'])

3600.0

avg_time_diff(['2023-05-28 12:00:00', '2023-05-28 13:00:00', '2023-05-28 14:00:00', '2023-05-28 15:00:00'])

3600.0

```python
from datetime import datetime

def avg_time_diff(dates: list[str]):
    if len(dates) <= 1:
        return 0.0
    differences = []
    for i in range(len(dates) - 1):
        first = datetime.strptime(dates[i], "%Y-%m-%d %H:%M:%S")
        second = datetime.strptime(dates[i + 1], "%Y-%m-%d %H:%M:%S")

        diff = second - first

        differences.append(diff.seconds)

    print(sum(differences) / len(differences)) 
```
