### **Найдите самую эффективную платформу по средним значениям конверсий**

````python
import csv


def get_best_metric_by_dimension(file_path, dimension_column, target_column):
    result = {}
    with open(file=file_path, mode="r", encoding="utf-8") as file:
        reader = csv.DictReader(file, delimiter=",")

        for row in reader:
            dimension = row[dimension_column]
            target = float(row[target_column])

            if dimension not in result:
                result[dimension] = {"sum": 0, "count": 0}

            result[dimension]["sum"] += target
            result[dimension]["count"] += 1
    averages = {dim: stats["sum"] / stats["count"] for dim, stats in result.items()}
    return max(averages, key=averages.get)


file_path = "data/campaign_data.csv"

most_successful_city = get_best_metric_by_dimension(file_path, "Город", "Конверсия")
most_effective_platform = get_best_metric_by_dimension(
    file_path, "Платформа", "Конверсия"
)
most_effective_platform
````

### **Напишите функцию, которая возвращает средний бюджет всех кампаний, запущенных на определенной платформе**

Функция `calculate_average_budget` должна принимать путь к файлу и название плафтормы и возвращать число - средний бюджет всех кампаний, запущенных на данной платформе, округленный до двух знаков после запятой.

````python
def calculate_average_budget(file_path: str, platform: str):
    with open(file=file_path, mode="r", encoding="utf-8") as file:
        dreader = csv.DictReader(file)

        total = 0
        counter = 0
        for row in dreader:
            if row["Платформа"] == platform:
                total += float(row["Бюджет"])
                counter += 1
        return round(total / counter, 2)


calculate_average_budget("data/campaign_data.csv", "Google")
````

### **Напишите генератор, который будет генерировать даты, в которые НЕ запускались кампании**

Функция `get_missing_campaign_dates` должна принимать путь к файлу и генерировать построчно даты, отсортированные по возрастанию, в которые не запускались кампании, в формате:

```
2022-02-29
2022-03-28
2022-03-29
2022-05-30
2022-06-13
2022-06-27
...
```

````python
def is_leap_year(year: int) -> bool:
    return (year % 4 == 0 and year % 100 != 0) or year % 400 == 0


def get_days_in_month(month: int, year: int) -> int:
    if month in (4, 6, 9, 11):
        return 30
    if month == 2:
        return 29 if is_leap_year(year) else 28
    return 31


def date_range_generator(start_str: str, end_str: str):
    year, month, day = map(int, start_str.split("-"))
    year_e, month_e, day_e = map(int, end_str.split("-"))

    while (year, month, day) <= (year_e, month_e, day_e):
        yield f"{year}-{month:02d}-{day:02d}"

        day += 1
        if day > get_days_in_month(month, year):
            day = 1
            month += 1
            if month > 12:
                month = 1
                year += 1
````

````python
def get_missing_campaign_dates(file_path):
    dates = set()
    with open(file_path, mode="r", encoding="utf-8") as file:
        dreader = csv.DictReader(file)

        for row in dreader:
            dates.add(row["Начальная дата"])

    for date_ in date_range_generator(min(dates), max(dates)):
        if date_ not in dates:
            yield date_


get_missing_campaign_dates("data/campaign_data.csv")
````

### **Напишите генератор, который будет генерировать данные о кампаниях, запущенных в указанном городе и имеющих бюджет выше заданного значения**

Функция `campaign_generator` должна принимать путь к файлу, город и бюджет и генерировать словари, отсортированные по возрастанию 'ID Кампании' в формате:

```
{'ID Кампании': '657', 'Тип': 'Контекстная', 'Платформа': 'Вконтакте', 'Доход': '10692'}
{'ID Кампании': '674', 'Тип': 'Нативная интеграция', 'Платформа': 'Яндекс', 'Доход': '20766'}
{'ID Кампании': '761', 'Тип': 'Соцсети', 'Платформа': 'Вконтакте', 'Доход': '15680'}
```

````python
def campaign_generator(file_path: str, target_city: str, target: int):
    with open(file_path, mode="r", encoding="utf-8") as file:
        dreader = csv.DictReader(file)

        for row in dreader:
            city = row["Город"]
            costs = float(row["Бюджет"])
            id = row["ID Кампании"]
            type_ = row["Тип кампании"]
            platform = row["Платформа"]

            if city == target_city and costs > target:
                yield {
                    "ID Кампании": id,
                    "Тип": type_,
                    "Платформа": platform,
                    "Доход": row["Доход"],
                }


campaign_generator("campaign_data.csv", "Москва", 25000)
````
