#### **Задача 1**

Задача не из легких, поэтому начнем по шагам. Для начала напишите функцию `run_tests`, которая прочтет эталонный код и код пользователя из `.py` файлов, выполнит их и сохранит результаты в файл `output.txt`.

**Решение**

Напишите свое решение ниже

```python
import importlib
```

```python

def run_tests(standard_code, user_code, test_cases, output_file):
    standart_module = importlib.import_module(standard_code)
    user_module = importlib.import_module(user_code)

    with open(file=output_file, mode="w", encoding="utf-8") as f:
        for case in test_cases:
            a, b = case
            result_from_etalon = str(standart_module.add(a, b))
            result_from_user = str(user_module.add_numbers(a, b))

            if result_from_etalon != result_from_user:
                f.write("Тест не пройден\n")
                f.write(f"Ожидаемый результат: {result_from_etalon}\n")
                f.write(f"Результат пользователя: {result_from_user}\n")
            else:
                f.write("Тест пройден\n")


# Чтение кодов из файлов
standard_module = "etalon"
user_module = "user"

test_cases = [(1, 2), (3, 4), (5, 6)]
output_file = 'output.txt'
run_tests(standard_module, user_module, test_cases, output_file)
```

Перед написанием кода, выполните следующую ячейку, в ней скачивается новый код пользователя, который проверит работу функции `check_for_malicious_code`.

**Решение**

Напишите свое решение ниже

```python
import inspect

FORBIDDEN_NAMES = ['eval', 'exec']


def check_for_malicious_code(code, output_file):
    for name in FORBIDDEN_NAMES:
        if name in code:
            return f"Код содержит вызов {name}()"
        

def run_tests(standard_code, user_code, test_cases, output_file):
    standart_module = importlib.import_module(standard_code)
    user_module = importlib.import_module(user_code)

    with open(file=output_file, mode="w", encoding="utf-8") as f:
        for case in test_cases:
            a, b = case
            result_from_etalon = str(standart_module.add(a, b))
            result_from_user = str(user_module.add_numbers(a, b))

            if result_from_etalon != result_from_user:
                f.write("Тест не пройден\n")
                f.write(f"Ожидаемый результат: {result_from_etalon}\n")
                f.write(f"Результат пользователя: {result_from_user}\n")
            else:
                f.write("Тест пройден\n")


# Чтение кодов из файлов
standard_module = "etalon"
user_module = "user_malicious"

test_cases = [(1, 2), (3, 4), (5, 6)]
output_file = 'output2.txt'
# Проверка на вредоносный код
malicious_code_message = check_for_malicious_code(
    inspect.getsource(importlib.import_module(user_module)), output_file
)

if malicious_code_message:
    with open(output_file, 'w') as f:
        f.write("Ошибка при проверке модуля user_malicious:\n")
        f.write(malicious_code_message)
else:
    run_tests(standard_module, user_module, test_cases, output_file)
```

```python
import time
import multiprocessing
```

```python

FORBIDDEN_NAMES = ['eval']

def check_for_malicious_code(code, output_file):
    pr = multiprocessing.Process(target=code)
    pr.start()

    pr.join(timeout=5)

    if pr.is_alive():
        pr.terminate()
        pr.join()
        with open(output_file, "w"):
            f.write("Предупреждение: время выполнения превышает 5 секунд.")


def run_tests(standard_code, user_code, test_cases, output_file):

    # сюда можете вставить ваш код из прошлой задачи
    # в него необходимо будет добавить проверку на время выполнения
    standart_module = importlib.import_module(standard_code)
    user_module = importlib.import_module(user_code)

    with open(file=output_file, mode="w", encoding="utf-8") as f:
        for case in test_cases:
            a, b = case
            result_from_etalon = str(standart_module.add(a, b))
            result_from_user = str(user_module.add_numbers(a, b))

            if result_from_etalon != result_from_user:
                f.write("Тест не пройден\n")
                f.write(f"Ожидаемый результат: {result_from_etalon}\n")
                f.write(f"Результат пользователя: {result_from_user}\n")
            else:
                f.write("Тест пройден\n")


# Чтение кодов из файлов
standard_module = "etalon"
user_module = "user_inf"

test_cases = [(1, 2), (3, 4), (5, 6)]
output_file = 'output3.txt'
# Проверка на вредоносный код
malicious_code_message = check_for_malicious_code(
    inspect.getsource(importlib.import_module(user_module)), output_file
)
if malicious_code_message:
    with open(output_file, 'w') as f:
        f.write("Ошибка при проверке модуля user_malicious:\n")
        f.write(malicious_code_message)
else:
    run_tests(standard_module, user_module, test_cases, output_file)
```
